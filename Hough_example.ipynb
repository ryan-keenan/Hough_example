```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#slope-intercept Hough example
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
m = np.linspace(0, 10, 100)
for x in range(10):
    y = x + np.random.random()*2
    ax1.scatter(x, y, s=500)
    b = y - m*x
    ax2.plot(m, b, lw=2)

ax2.set_xlim(0,3)
ax2.set_ylim(-10,11)
ax1.tick_params(axis='both', which='major', labelsize=20)
ax1.set_xlabel('x', fontsize=50)
ax1.set_ylabel('y', fontsize=50, rotation='horizontal')
ax1.set_title('Image Space', fontsize=50)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xlabel('m', fontsize=50)
ax2.set_ylabel('b', fontsize=50, rotation='horizontal')
ax2.set_title('Hough Space', fontsize=50)
```

```python
#polar coords Hough example
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
theta = np.linspace(-0.5, 3, 100)
for x in range(10):
    y = 10-x
    ax1.scatter(x, y, s=500)
 
    rho1 = x*np.cos(theta) + y*np.sin(theta)
    ax2.plot(theta, rho1, lw=2)

ax2.set_xlim(0,3)
ax2.set_ylim(-10,11)
ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.set_xlabel('x', fontsize=50)
ax1.set_ylabel('y', fontsize=50, rotation='horizontal')
ax1.set_title('Image Space', fontsize=50)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xlabel(r'$\theta$', fontsize=50)
ax2.set_ylabel(r'$\rho$', fontsize=50, rotation='horizontal')
ax2.set_title('Hough Space', fontsize=50)
```

```python
#building a square
bottomx = np.linspace(1, 5, 9)
leftx = np.ones(7)
xsquare = np.concatenate((bottomx, leftx, bottomx, leftx+4), axis=0)
bottomy = np.ones(9)
lefty = np.linspace(1.5, 4.5, 7)
ysquare = np.concatenate((bottomy, lefty, bottomy+4, lefty), axis=0)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.scatter(xsquare, ysquare, s=500)

theta = np.linspace(1, 4, 100)
for idx, x in enumerate(xsquare):
    y = ysquare[idx]    
    rho = x*np.cos(theta) + y*np.sin(theta)
    ax2.plot(theta, rho, lw=2)
    
ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.set_xlabel('x', fontsize=50)
ax1.set_ylabel('y', fontsize=50, rotation='horizontal')
ax1.set_title('Image Space', fontsize=50)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xlabel(r'$\theta$', fontsize=50)
ax2.set_ylabel(r'$\rho$', fontsize=50, rotation='horizontal')
ax2.set_title('Hough Space', fontsize=50)
```
